```python
def factorial(number):
  if number == 1:
    return number
  elif number < 1:
    return None
  else
    return number * factorial(number - 1)

factorial(5)
```

```python
def factorial(number):
  answer = 1
  for n in range(number, 0, -1):
    answer *= n

  return answer

factorial(5)
```
