# Finding the Noise Makers


````python
def add_to_noise_makers(name, file='noise_makers.txt'):
  # Adds to the noise makers file
  with open(file, 'w') as file:
    file.write(name + '\n')

number_of_noise_makers = int(input("How many people made noise? "))

noise_makers = []
valid_noise_makers = []

for i in range(number_of_noise_makers):
  noise_maker = input("Enter name: ")
  noise_makers.append(noise_maker.lower())

with open('students.txt') as students:
  student = students.readline()
  while student != '':
    student = student.strip('\n')
    if student.lower() in noise_makers:
      add_to_noise_makers(student)
      valid_noise_makers.append(student.lower())
    student = students.readline()

print(f"Valid noise makers: {valid_noise_makers}")
print(f"Your noise makers: {noise_makers}")
print('Conditions')

if len(noise_makers) == len(valid_noise_makers):
  print("All noise makers are valid")
else:
  print("Invalid noise maker found:")
  print(list(set(noise_makers) - set(valid_noise_makers)))
````

# CSV and JSON
CSV stands for comma seperated values
```
Students,Class,Age
John,SS2,12
Mary,Grade 5,6
```

We use the `csv` library to read CSV files in python. The four main operations we can open, close, read and write. Opening and closing CSV files is the same as opening and closing any other file but reading and writing is different

To write to a csv file you use the `writer()` function

````python
import csv
with open('test.csv', 'w') as file:
  writer = csv.writer(file)
  writer.writerow(['Students', 'Class', 'Age'])
  writer.writerow(['John', 'SS2', '12'])
````

````python
with open('test.csv', 'a') as file:
  writer = csv.writer(file)
  writer.writerow(['Mary', 'Grade 4', '8'])
````

To read a csv file we used the `reader()` function

````python
with open('test.csv', 'r') as file:
  reader = csv.reader(file)
  header_rows = next(reader)
  print(header_rows)
  for row in reader:
    print(row)
````

````python
with open('test.csv', 'r') as file:
  reader = csv.reader(file)
  header_rows = next(reader)
  print(header_rows)
  for row in reader:
    print(row)
  print(reader.line_num)
````

````python
students = [['Matthew', 'JSS1', 13], ['Cynthia', 'SS3', '17']]
with open('test.csv', 'a') as file:
  writer = csv.writer(file)
  writer.writerows(students)
````

````python
students = [{'Students': 'Moses',
             'Class': 'Grade 3',
             'Age': 10
             },
            {
                'Students': 'Rebecca',
                'Class': 'Grade 5',
                'Age': 8
            }]
fields = list(students[0].keys())
with open('test.csv', 'a') as file:
  writer = csv.DictWriter(file, fieldnames=fields)
  writer.writeheader()
  writer.writerows(students)
````

To read rows into a dictionary we use `DictReader()` function

````python
with open('test.csv', 'r') as file:
  reader = csv.DictReader(file)
  for row in reader:
    print(row)
````

````python
with open('test.csv', 'r') as file:
  reader = csv.DictReader(file)
  for row in reader:
    print(row['Students'])
````

# What is JSON?
JSON stands for JavaScript Object Notation.

We use the JSON library to write TO JSON fiLES.

To read info we use `json.load` and to write info we use `json.dump`

````python
import json

with open('sample.json', 'r') as file:
  data = json.load(file)
  print(data)
````

````python
info_to_add = {
  "name": "Mark Henry",
  "age": 24
}

with open("sample.json", "w") as file:
  json.dump(info_to_add, file)
````
