## Variables

Can only have one data type associated with it


Consists of:
- a name
- an assignment operator (=)
- and a value

Can be named as
- Snake case (`snake_case`): using underscore in place of spaces
- Camel case (`camelCase`): capitalizing the first letter of the next words
- Pascal case (`PascalCase`): capitalizing the first letter of each word.

```python
name = "Tosin"
age = 15
```

```python
# camel case
numberOfCars = 2
numberOfCars
```

```python
# pascal case
NumberOfCars = 34
NumberOfCars
```

```python
# snake case
number_of_cars = 22
number_of_cars
```

- A variable must not end with a full stop or any special characters
- You cannot print a variable without creating it beforehand

```python
print("NumberOfCars")  # doesn't print the variable value
```
