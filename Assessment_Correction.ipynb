# Runner up score
Given the participants' score sheet for your University Sports Day, you are required to find the runner-up score. You are given _n_ scores. Store them in a list and find the score of the runner-up.

### Input Format
The first line contains _n_. The second line contains an array _A[]_ of _n_ integers each separated by a space.

#### Output Format

Print the runner-up score.

> Sample Input 0
```
5
2 3 6 6 5
```
> Sample Output 0
```
5
```

Explanation 0

Given list is `[2,3,6,6,5]`. The maximum score is 6, second maximum is 5. Hence, we print 5 as the runner-up score.

````python
# if __name__ == '__main__': # Line not needed for google colab
# Default code for collecting the input
n = int(input())
arr = map(int, input().split())
arr = list(arr) # converting to a list

max_score = arr[0]
for score in arr:
    if score > max_score:
        max_score = score

# max_score = max(arr)
# print(max_score)


scores = set(arr)
scores.remove(max_score)
scores = list(scores)
runner_up = scores[0]
for score in scores:
    if score > runner_up:
        runner_up = score

print(runner_up)
````

# Election Result Integrity
After a chaotic national election, the electoral commission hires you to write a program to validate polling unit submissions, detect suspicious patterns, and compute final candidate tallies.

Each polling unit sends its results as a tuple in the form:

`(unit_id, candidate_name, votes)`

Where:

    unit_id → a string
    candidate_name → a string
    votes → a string or int (because some units submit votes as strings — because, well… Nigeria.)

Your task is to validate all entries, detect anomalies, and compute the final election table.

Use this as your input
```py
entries = [
    ("PU001", "Adeola", "230"),
    ("PU001", "Bala", 500),
    ("PU002", "Adeola", "hello"),
    ("PU003", "Chioma", 720),
    ("PU004", "Bala", 400),
    ("PU004", "Adeola", 300),
    ("PU004", "Bala", 200),
    ("PU005", "Tunde", 150),
    ("PU005", "Latifah", 200),
    ("PU006", "Samuel", -5),
    ("PU006", "Kemi", 180),
    ("PU007", "Ibrahim", 800)
]
```
**Hint**: You can use the `sorted()` function to sort a dictionary

### Input Format

entries — a list of tuples, size `N`

Each tuple is of the form:

`(unit_id, candidate_name, votes)`

> Example Input:
```py
entries = [
    ("PU001", "Adeola", "230"),
    ("PU001", "Bala", 500),
    ("PU002", "Adeola", "hello"),   # invalid votes
    ("PU003", "Chioma", 720),       # suspicious: votes > 700
    ("PU004", "Bala", 400),
    ("PU004", "Adeola", 300),
    ("PU004", "Bala", 200),         # duplicate candidate in same PU
]
```
> Expected Output (for example)
```py
{
  "valid_units": 2,
  "invalid_entries": 2,
  "suspicious_units": {"PU003"},
  "results": {
    "Bala": 900,
    "Adeola": 530,
    "Chioma": 720
  }
}
```
### Rules & Requirements

    Validation A transaction is invalid if:
        * unit_id is empty or not a string
        * candidate_name contains digits (e.g., "Peter5")
        * votes cannot be converted to a positive integer
        * votes is negative
        * any field is missing or None
        * a polling unit reports the same candidate multiple times (duplicates)

    Anomaly Detection Mark a unit as “SUSPICIOUS” if:
        * total votes across all candidates in that unit exceed 1000
        * votes for any candidate exceed 700 (statistical * impossibility per new INEC guidelines)
        * the unit reports only one candidate (flag as possible ballot-stuffing)

Output Format

You must return a dictionary with the following keys:
```py
{
    "valid_units": <number of valid polling units>,
    "invalid_entries": <number of invalid tuples>,
    "suspicious_units": <set of unit_ids flagged>,
    "results": {
        candidate_name: total_votes,
        ...
    }
}
```

results must be sorted by descending votes

````python
entries = [
    ("PU001", "Adeola", "230"),
    ("PU001", "Bala", 500),
    ("PU002", "Adeola", "hello"), # text
    ("PU003", "Chioma", 720), # suspicious
    ("PU004", "Bala", 400),
    ("PU004", "Adeola", 300),
    ("PU004", "Bala", 200), # duplicate
    ("PU005", "Tunde", 150),
    ("PU005", "Latifah", 200),
    ("PU006", "Samuel", -5), # negative
    ("PU006", "Kemi", 180),
    ("PU007", "Ibrahim", 800) # suspicious
]

def is_valid(entry):
    votes = str(entry[2])
    numbers = '0123456789'
    for character in votes:
        if character in numbers:
            continue
        else:
            return False

    if int(votes) < 0:
        return False

    for item in entry:
        if item == "" or item is None:
            return False

    return True

def is_duplicate(entries, unit_id, candidate):
    for entry in entries:
        if unit_id == entry[0] and candidate == entry[1]:
            return True
    return False

accepted_entries = []  #
no_invalid = 0

for entry in entries:
    if is_valid(entry):
        current_id = entry[0]
        current_candidate = entry[1]
        if is_duplicate(accepted_entries, current_id, current_candidate):
            no_invalid += 1
        else:
            accepted_entries.append(entry)
    else:
        no_invalid += 1

# print(accepted_entries)

def check_suspicious(entries):
    suspicious_entries = set([])
    for entry in entries:
        if int(entry[2]) > 700:
            suspicious_entries.add(entry[0])

    # checking ids available
    unit_ids = set([])
    for entry in entries:
        unit_ids.add(entry[0])
    #checking submission per id
    for unit_id in list(unit_ids):
        submissions = 0
        total_votes = 0

        for entry in entries:
            if unit_id == entry[0]:
                submissions += 1
                total_votes += int(entry[2])

        if submissions == 1:
            suspicious_entries.add(unit_id)
        elif total_votes > 1000:
            suspicious_entries.add(unit_id)
    return suspicious_entries

sus_entries = check_suspicious(accepted_entries)
# print(sus_entries)

candidates = set([])
for entry in accepted_entries:
    candidates.add(entry[1])
candidates = list(candidates)


def count_votes(entries, candidate):
    total_votes = 0
    for entry in entries:
        if entry[1] == candidate:
            total_votes += int(entry[2])
    return total_votes

results = {}
for candidate in candidates:
    results[candidate] = count_votes(accepted_entries, candidate)
# print(results)

results = {k: v for k, v in sorted(results.items(), key=lambda item: item[1], reverse=True)}

finals = {
    "valid_units": len(accepted_entries),
    "invalid_entries": no_invalid,
    "suspicious_units": sus_entries,
    "results": results
}
print(finals)
````

# Nested Lists
Given the names and grades for each student in a class of students, store them in a nested list and print the name(s) of any student(s) having the second lowest grade

````python
students = []
grades = []
for _ in range(int(input())):
    name = input()
    score = float(input())
    grades.append(score)
    student = [name, score]
    students.append(student)

lowest_grade = grades[0]
for grade in grades:
    if grade < lowest_grade:
        lowest_grade = grade
# print(lowest_grade)

grades = set(grades)
grades.remove(lowest_grade)
grades = list(grades)
second_lowest_grade = grades[0]
for grade in grades:
    if grade < second_lowest_grade:
        second_lowest_grade = grade
# print(second_lowest_grade)

students_second_lowest = []
for student in students:
    name = student[0]
    score = student[1]
    if score == second_lowest_grade:
        students_second_lowest.append(name)

for name in sorted(students_second_lowest):
    print(name)
````
