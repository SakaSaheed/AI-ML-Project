## Slicing and Indexing
To slice a sequence you use this format
```py
sequence[start_index:stop_index:step]
```
Each of the options `start_index`, `stop_index` and `step` are optional.

````python
name = "Pythony"
````

````python
name[::]
````

````python
name[2:]
````

````python
name[2::2]
````

````python
name[2::-1]
````

````python
name = "Python"
name[-6::2]
````

````python
name[::-1]
````

````python
name[2:]
````

````python
name[-6::-1]
````

````python
name[0::-1]
````

````python
name[2::-1]
````

## Tuples
Tuples are another sequential type similar to lists but the difference is that tuples are immutable. They can be made using the `()`.

````python
days_of_the_week = ('Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday')
del days_of_the_week
days_of_the_week # Will give error
````

You can unpack tuples by attaching them to variables. Typically the number of variables must be the exact same thing as the number of items in the tuples but you can unpack multiple items in one variable by using the `*`.

````python
christmas_gifts = ('chicken', 'rice', 'groundnut oil', 'garri', 'ps5')
*kunle, shayo, matthew, cruz = christmas_gifts
kunle
````

## Membership operators
These are used to check if an item exists in a collection/sequence. It has only two operators:
- `in`

````python
name = input("Input your name: ")
if '1' in name:
  print('There is a number')
````

````python
invites_list = ['Hustler', 'Yahoo Boy', 'Gambler', 'Kidnapper']
name = 'Everyday man'
if name not in invites_list:
  print("Go and pay for the program")
````

## Loops
There are two types of loops `for` and `while` loops. There are used when you want to repeat a section of code based on a condition or a particular number of times.

````python
fruits = ['apple', 'banana', 'orange']
for fruit in fruits:
  print(fruit)
````

````python
manifesto = "build roads"
while manifesto == "build roads":
  print("emi lokan")
````

````python
tenure = 0
while tenure < 2:
  print("keep campaigning")
  tenure += 1
````

````python
for number in range(1,11):
  print(number)
````

````python
for i in range(2):
  print(i)
````

````python
fruits = ['apple', 'banana', 'orange']
# for fruit in fruits:
#   print(fruit)

for index in range(len(fruits)):
  print(fruits[index])
````

## `range` data types
We use `range` when we want to make a sequence of numbers. You can make a range using the range function in this format:
```py
range(start_number, stop_number, step)
```
All of the arguments except `stop_number` is optional so you can write range like this `range(stop_number)` and it will start counting from 0

````python
for i in range(5, 51, 5):
  print(i)
````

````python
for i in range(10, 0, -1):
  print(i)
````

````python
password = '1234'
for i in range(3):
  user_password = input("Enter your password: ")
  if user_password == password:
    print("Access granted")
    break
  else:
    print("Access denied")
  if i == 2:
    print("Times up")
print("Continuation")
````

````python
for i in range(3):
  for j in range(3):
    print('*', end="")
  print()
````

````python
sequence = range(3)
for number in sequence:
  print(number)
````
