```python
team = {
    'Messi': 10,
    'Ronaldo': 7,
    'Neymar': 11,
    'Osimhen': 9
}

new_gen = {
    'Haaland': 10,
    'Mbappe': 7,
    'Salah': 11,
    'Lukaku': 9,
    'Palmer': 10,
    'Yamal': 10
}

team.update(new_gen)
print(list(team.keys()))
```

```python
player_names = list(team.keys())
for name in player_names:
    print(f'An award for {name}')
```

```python
player_numbers = list(team.values())
for number in player_numbers:
    print(f'An award for {number}')
```

```python
nigerian_team = {
    'Osimhen': 9,
    'Musa': 10,
    'Lookman': 11,
    'Iwobi': 7,
    'Bassey': 3,
    'Ekong': 8,
    'Nwambali': 33,
    'Boniface': 4
}

def measure_team_strength(team):
  base_score = 4
  if team.get('Osimhen'):
    base_score += 4
  elif team.get('Nwambali'):
    base_score += 3
  elif team.get("Lookman"):
    base_score -= 2
  return base_score

print(measure_team_strength(nigerian_team))
```
